## Importações

```python
import os
import win32com.client
import pandas as pd
from datetime import datetime
```

## Leitura do email

```python
# Configurações
remetente_amigavel = "Transportes"  # Nome exibido do remetente
assunto_email = "RCA PALETIZAÇÃO NOVEMBRO 2"
caminho_arquivo = os.path.join('C:\\', 'arquivo.xlsx')
nome_pasta_outlook = "RCA"
```

```python
# Conectar ao Outlook
try:
    outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")
    pasta = outlook.Folders.Item("").Folders.Item(nome_pasta_outlook)
    print(f"Pasta '{nome_pasta_outlook}' acessada com sucesso.")
except Exception as e:
    print(f"Erro ao acessar a pasta '{nome_pasta_outlook}': {e}")
    exit()
```

```python
# Buscar emails na pasta
try:
    emails = pasta.Items
    emails.Sort("[ReceivedTime]", True)
    print("Emails carregados com sucesso.")
except Exception as e:
    print(f"Erro ao carregar emails da pasta: {e}")
    emails = None
```

```python
# Procurar email específico e salvar o anexo
pasta_destino = r"C:"
email_encontrado = None
if emails:
    print("\nListando emails na pasta:")
    for email in emails:
        print(f"Assunto: {email.Subject}, Remetente: {email.SenderName}, Data: {email.ReceivedTime}")
        if assunto_email in email.Subject and remetente_amigavel in email.SenderName:
            email_encontrado = email
            break

if email_encontrado:
    print("\nEmail encontrado!")
    for anexo in email_encontrado.Attachments:
        if anexo.FileName.endswith(".xlsx"):
            caminho_arquivo = os.path.join(pasta_destino, anexo.FileName)
            anexo.SaveAsFile(caminho_arquivo)
            print(f"Anexo salvo em: {caminho_arquivo}")
else:
    print("\nEmail não encontrado com o assunto e remetente especificados.")
    exit()
```
